 # Categorical plots: This plot deal with categorical variables and how they can be visualized.

```python
# import the seaborn library
import seaborn as sns
#import done to avoid warnings
from warnings import filterwarnings
#READING the dataset
df=sns.load_dataset('tips')
#first five entries if the dataset
df.head()
```

# BARPLOT:- 
        to use this plot we choose a categorical column for the x axis and a numerical column for the y axis and we see that 
it creates a plot taking a means per categorical column.

```python
#Syntax:
# barplot([x,y,hue,data,order,,hue_order,..])
#set the background style of the plot
sns.set_style('darkgrid')
#plot the graph using the default estimator mean
sns.barplot(x='sex',y='total_bill',data=df,palette='plasma')
#or
import numpy as np
#change the estimator from means to standard devistion
sns.barplot(x='sex',y='total_bill',data=df, palette='plasma',estimator=np.std)
```

# Vilolinplot

estimation to given a better description about the data distrubution.

```python
import seaborn as sns
df=sns.load_dataset('tips')
sns.violinplot(x='day',y='total_bill',data=df,hue='sex',split=True)
```

```python
#Stripplot:
   # It basically create a scatter plot based on the category.
sns.stripplot(x='day',y='total_bill',data=df,jitter=True,hue='smoker',dodge=True)
```

# Pairplot In Seaborn:
        In Seaborn, pairplot is a powerful function that allows you to visualize the relationships between multiple variables 
        in a dataset, typically through a grid of scatterplots and histograms or kernel density estimates (KDE) on the diagonal. 
        It can help you explore the pairwise correlations between continuous variables and identify patterns such as clusters,
        outliers, or linear relationships.

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Load a sample dataset (e.g., Iris dataset)
df = sns.load_dataset('iris')
# Create a pairplot of the entire dataset
sns.pairplot(df)
plt.show()
```

```python
# Create a pairplot with coloring by species and regression lines
sns.pairplot(df, hue='species', kind='reg', height=2.5)
plt.show()
```

```python
# Create a pairplot with KDE on the diagonal and scatter on off-diagonal
sns.pairplot(df, diag_kind='kde', height=2.5)
plt.show()
```

```python
# Create a pairplot with different markers for each species
sns.pairplot(df, hue='species', markers=["o", "s", "D"], height=2.5)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# If you're using a Jupyter notebook, use the magic function %matplotlib inline to display plots inline
%matplotlib inline  

# Load the tips dataset
dataset = sns.load_dataset('tips')

# Show the first five entries of the dataset
print(dataset.head())

numeric_dataset=dataset.select_dtypes(include=['float64','int64'])


# Compute the correlation between the different parameters
tc =numeric_dataset.corr()  # correlation asign numberic value only.

# Plot a heatmap of the correlation data
sns.heatmap(tc, annot=True, cmap='coolwarm', fmt='.2f', cbar=True)

# Show the plot
plt.show()
```

# Cluster maps:-
  Cluster maps use hierarchical clustering.it performs the clustering based on the similarity of the woes and columns.

```python
#import the necessary libraries:
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# If you're using a Jupyter notebook, use the magic function %matplotlib inline to display plots inline
%matplotlib inline  
#load the flights dataset:
fd=sns.load_dataset('flights')
#make a dataframe of the data
df=pd.pivot_table(values='passengers',index='month',columns='year',data=fd)
#first five entries of the dataset
df.head()
#make a clustermap from the  dataset
sns.clustermap(df,cmap='plasma')
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# If you're using a Jupyter notebook, use the magic function %matplotlib inline to display plots inline
%matplotlib inline  

fd=sns.load_dataset('flights')
df=pd.pivot_table(values='passengers',index='month',columns='year',data=fd)
df.head()
sns.clustermap(df,cmap='plasma',standard_scale=1)
```
