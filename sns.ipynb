# Titanic Datasets:

```python
import numpy as np
import pandas as pd
import seaborn as sns

titanic=pd.read_csv('tested.csv')
titanic
```

```python
sns.set_style('darkgrid')
sns.barplot(x='Sex',y='Name',data=titanic,palette='plasma')
import numpy as np
sns.barplot(x='Sex',y='Name',data=titanic, palette='plasma',estimator=np.std)
```

```python
sns.stripplot(x='Age',y='Pclass',data=titanic,jitter=True,hue='Sex',dodge=True)
```

```python
sns.swarmplot(x='Sex',y='Age',data=titanic)
```

```python
sns.violinplot(x='Sex',y='Age',data=titanic)
sns.swarmplot(x='Sex',y='Age',data=titanic,color='black')
```

```python
sns.factorplot(x='Sex',y='Age',data=titanic,kind='bar')
```

```python
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
sns.set_style('whitegrid')
sns.distplot(titanic['Pclass'], kde = False, color='red', bins=30)
```

```python
sns.jointplot(x='Pclass',y='Sex',data=titanic)	
```

```python
sns.jointplot(x='Age',y='Pclass',data=titanic,kind='kde')	
```

```python
sns.set_style('whitegrid')
sns.lmplot(x='Age',y='Pclass',data=titanic)	
```

```python
sns.set_style('whitegrid')
sns.lmplot(x='Age',y='Pclass',data=titanic,hue='Sex',markers=['o','v'])	
```

```python
sns.set_style('whitegrid')
sns.lmplot(x='Age',y='Pclass',data=titanic,hue='Sex',markers=['o','v'],scatter_kws={'s':100},palette='plasma')	
```

```python
sns.lmplot(x='Age',y='Pclass',data=titanic,col='Sex',row='Sex',hue='Sex')	
```

```python
sns.lmplot(x='Age',y='Pclass',data=titanic,col='Embarked',row='Sex',hue='Sex')	
```

```python
sns.lmplot(x='Age',y='Pclass',data=titanic,col='Embarked',row='Sex',hue='Sex',aspect=0.6,palette='coolwarm')
```
