# Box Plot:

```python
  A Box Plot,also known as a whisker plot,is a standardized way of displaying the distribution of data
based on a five-number summary:minimum,first qurtile(Q1), median(Q2),third quartile(Q3),and maximum.
  It can also show ourliers. Lets see an example of how to create a Box Plot using matplotlib in python:
```

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(10)
data=[np.random.normal(0,std,100) for std in range(1,4)]
#creare a box plot
plt.boxplot(data,
            vert=True,
            patch_artist=True,
            boxprops=dict(facecolor='skyblue'),
            medianprops=dict(color='red'))
plt.xlabel('Data Set')
plt.ylabel('Values')
plt.title('Example of Box plot')
plt.show()
```

# Heatmap:

```python
  A Heatmap is a data visulization technique that represents data in a matrix form,where individual values are
represented as colors. Heatmaps are particulrly useful for visualizing the magnitude of data across a two-dinmensional
surfacce and identifying patterns,correlations and concentrations.
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(0)
data=np.random.rand(10,10)
#create a heatmap
plt.imshow(data,cmap='viridis',
           interpolation='nearest')
#add a color bar to show the scale
plt.colorbar()
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Example of Heatmap')
plt.show()
```

# Subplot2grid()

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=[20,25,35,55]
axes1=plt.subplot2grid((7,1),(0,0),rowspan=2,colspan=1)
axes1.plot(x,y)

axes2=plt.subplot2grid((7,1),(0,0),rowspan=2,colspan=1)
axes2.plot(y,x)
```
