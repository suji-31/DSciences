```python
import numpy as np
img=cv2.imread('sun.jpg')
cv2.imshow('Original Image',img)

g=cv2.GaussianBlur(img,(7,7),0)
cv2.imshow('Gaussian Blur',g)

m=cv2.medianBlur(img,5)
cv2.imshow('Median Blur',m)

b=cv2.bilateralFilter(img,9,75,75)
cv2.imshow('Bilateral Blur',b)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
image=cv2.imread('sun.jpg')
B, G, R = cv2.split(image)
# Corresponding channels are separated
cv2.imshow("original",image)
cv2.waitKey(0)
cv2.imshow("blue",B)
cv2.waitKey(0)
cv2.imshow("Green",G)
cv2.waitKey(0)
cv2.imshow("red",R)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
